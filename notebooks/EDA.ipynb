```python
import pandas as pd
import plotly.express as px
import streamlit as st

car_data = pd.read_csv('vehicles_us.csv') # leer los datos
```

```python
display(car_data.head())
car_data.info()
```

```python
# Calcular la media o mediana de las columnas relevantes para rellenar los valores nulos
model_year_mean = car_data['model_year'].mean()
cylinders_median = car_data['cylinders'].median()
odometer_median = car_data['odometer'].median()

# Rellenar valores nulos con asignación directa
car_data['model_year'] = car_data['model_year'].fillna(model_year_mean)
car_data['cylinders'] = car_data['cylinders'].fillna(cylinders_median)
car_data['odometer'] = car_data['odometer'].fillna(odometer_median)
car_data['paint_color'] = car_data['paint_color'].fillna('unknown')
car_data['is_4wd'] = car_data['is_4wd'].fillna(0)

# Convertir 'model_year' a tipo int
car_data['model_year'] = car_data['model_year'].astype('int')

# Convertir 'is_4wd' a 1 y 0 (de entero) antes de reemplazar
car_data['is_4wd'] = car_data['is_4wd'].apply(lambda x: 1 if x > 0 else 0)

# Convertir 'is_4wd' a "Sí" y "No"
car_data['is_4wd'] = car_data['is_4wd'].replace({1: 'Sí', 0: 'No'})

# Convertir 'date_posted' a tipo datetime
car_data['date_posted'] = pd.to_datetime(car_data['date_posted'])

display(car_data.head())
car_data.info()
```
